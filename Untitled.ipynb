```python
#라이브러리 선언
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
import pandas as pd
import os
import cv2
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
```

```python
path = 'C:/users/wjddu/classification/'   # 전체경로
```

```python
def add_noise(img):
    row, col, _ = img.shape

    number_of_pixels = np.random.randint(3, 28)
    
    for i in range(number_of_pixels):
        y_coord=np.random.randint(0, row)
        x_coord=np.random.randint(0, col)
        img[y_coord][x_coord][:] = 255
        
    number_of_pixels = np.random.randint(3, 28)
    
    for i in range(number_of_pixels):
        y_coord=np.random.randint(0, row)
        x_coord=np.random.randint(0, col)
        img[y_coord][x_coord][:] = 0
        
    return img
```

```python
def flip(img):
    img = cv2.flip(img, -1)
    return img
```

```python
img_list_np = []                # 이미지를 모두 담을 배열 선언
img_list = ['chicken', 'pizza', 'hamburger', 'ramen', 'pork']
for j in img_list:
    filepath = os.listdir(path+j+'/')
    print(path+j+'/')
    for i in range(len(filepath)):
        img_path = path+j+'/'+filepath[i]
        img = cv2.imread(img_path)     
        img = cv2.resize(img, (28, 28))                    
        img_array = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)    
        img_list_np.append(img_array)                       
        img_list_np.append(flip(img_array))
        img_list_np.append(add_noise(img_array))

train_images = np.array(img_list_np) #리스트를 numpy로 변환
```

```python
train_labels = []                                                  # 훈련 데이터에 대한 라벨을 담을 리스트 선언
class_names = ['chicken', 'pizza', 'hamburger', 'ramen', 'pork']   # 라벨에 따른 클래스의 이름 선언
for i in range(5):
    for j in range(90):
        train_labels.append(i)                                     # 0,1,2,3,4를 30개씩 라벨 리스트에 담기 
train_labels = np.array(train_labels)      
```

```python
print("train_labels : ", train_labels[0],                               # 트레이닝 라벨이 잘 들어갔는지 확인
      ",  train_labels : ", class_names[train_labels[0]])
print("train_labels : ", train_labels[360], 
      ",  train_labels : ", class_names[train_labels[360]])
```

# L2 정규화

```python
train_images = train_images.astype('float64')
train_images = tf.math.l2_normalize(x = train_images, axis=1)
```

```python
model = Sequential([
    layers.Conv2D(32, (2,2), padding='same', strides=(1,1), activation='relu', input_shape=(28,28,3)),
    layers.MaxPooling2D(pool_size=(2,2)),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    layers.Conv2D(64, (2,2), padding='same', strides=(1,1), activation='relu'),
    layers.MaxPooling2D(pool_size=(2,2)),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(5, activation='softmax') 
])
model.compile(optimizer='adam', loss = 'sparse_categorical_crossentropy', metrics = 'accuracy')
history = model.fit(train_images, train_labels, epochs=20, batch_size=6)
```

```python
pre = model.predict(train_images)
```

```python
for i in range(len(pre)):
    if i%90 ==0:
        print("--------------------------------------------------")
    print(np.argmax(pre[i]))
```
